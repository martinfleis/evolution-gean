Get the initial table of case studies with the centre coordinate for a buffer area.

Not a part of reproducible code, will not be published.

```python
import geopandas as gpd
import pandas as pd
import glob
import pathlib
```

```python
files = glob.glob("../../Dropbox/Academia/Data/Geo/Resilience/data_resilience_e_ais/*/*/data.gpkg")
```

```python
cases = pd.DataFrame(columns=["period", "origin"])

for f in files:
    path = pathlib.Path(f)
    buffer = gpd.read_file(f, layer='buffer400').to_crs(4326)
    point = list(buffer.geometry.iloc[0].centroid.coords)[0]
    cases.loc[path.parent.name] = [path.parent.parent.name, point]
```

```python
cases
```

```python
cases = cases.reset_index()
```

```python
cases = cases.rename(columns={"index": "case"})
```

```python
cases = cases.replace({"1_pre-Industrial": "pre-industrial", "2_industrial": "industrial", "3_Modernist": "modernist", "4_Garden_city": "garden city", 
                      "5_Neo-traditional": "neo-traditional", "6_Informal": "informal"})
```
